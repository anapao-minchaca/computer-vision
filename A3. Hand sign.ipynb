### Imports

```python
import cv2 
import os, glob

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

image_dir = "images/hand_gestures_digits_0_5"
```

### Plot all the base images in a subplot of 2 rows and 3 columns

```python
## Code to plot the base images.

mano_cerrada = cv2.imread("images/hand_gestures_digits_0_5/01.jpg")
mano_cerrada = cv2.cvtColor(mano_cerrada, cv2.COLOR_BGR2RGB)

uno = cv2.imread("images/hand_gestures_digits_0_5/02.jpg")
uno = cv2.cvtColor(uno, cv2.COLOR_BGR2RGB)

dos = cv2.imread("images/hand_gestures_digits_0_5/03.jpg")
dos = cv2.cvtColor(dos, cv2.COLOR_BGR2RGB)

tres = cv2.imread("images/hand_gestures_digits_0_5/04.jpg")
tres = cv2.cvtColor(tres, cv2.COLOR_BGR2RGB)

cuatro = cv2.imread("images/hand_gestures_digits_0_5/05.jpg")
cuatro = cv2.cvtColor(cuatro, cv2.COLOR_BGR2RGB)

cinco = cv2.imread("images/hand_gestures_digits_0_5/06.jpg")
cinco = cv2.cvtColor(cinco, cv2.COLOR_BGR2RGB)

f3, ((ax31, ax32, ax33), (ax34, ax35, ax36)) = plt.subplots(2, 3, figsize=(10,10))
ax31.set_title("0")
ax31.imshow(mano_cerrada)
ax32.set_title("1")
ax32.imshow(uno)
ax33.set_title("2")
ax33.imshow(dos)

ax34.set_title("3")
ax34.imshow(tres)
ax35.set_title("4")
ax35.imshow(cuatro)
ax36.set_title("5")
ax36.imshow(cinco)
```

### Process all the images. Add all the code cells you may require.

```python
## Image processing and resize
cerrada_gray = cv2.cvtColor(mano_cerrada, cv2.COLOR_RGB2GRAY)
resize1 = cv2.resize(cerrada_gray, (400, 520))

uno_gray = cv2.cvtColor(uno, cv2.COLOR_RGB2GRAY)
resize2 = cv2.resize(uno_gray, (400, 520))

dos_gray = cv2.cvtColor(dos, cv2.COLOR_RGB2GRAY)
resize3 = cv2.resize(dos_gray, (400, 520))

tres_gray = cv2.cvtColor(tres, cv2.COLOR_RGB2GRAY)
resize4 = cv2.resize(tres_gray, (400, 520))

cuatro_gray = cv2.cvtColor(cuatro, cv2.COLOR_RGB2GRAY)
resize5 = cv2.resize(cuatro_gray, (400, 520))

cinco_gray = cv2.cvtColor(cinco, cv2.COLOR_RGB2GRAY)
resize6 = cv2.resize(cinco_gray, (400, 520))

## Filtros
kernel = np.ones((3,3), np.uint8)
images = [resize1, resize2, resize3, resize4, resize5, resize6]

for i in range(len(images)):
    blur = cv2.GaussianBlur(images[i], (9,9), 0)
    edges = cv2.Canny(blur, 20, 50)
    dilation = cv2.dilate(edges, kernel, iterations = 1)
    images[i] = dilation
```

### Plot all the processed images in a subplot of 2 rows and 3 columns.

```python
## Code to plot the processed images.

f3, ((ax31, ax32, ax33), (ax34, ax35, ax36)) = plt.subplots(2, 3, figsize=(10,10))
ax31.set_title("0")
ax31.imshow(images[0], cmap="gray")
ax32.set_title("1")
ax32.imshow(images[1], cmap="gray")
ax33.set_title("2")
ax33.imshow(images[2], cmap="gray")

ax34.set_title("3")
ax34.imshow(images[3], cmap="gray")
ax35.set_title("4")
ax35.imshow(images[4], cmap="gray")
ax36.set_title("5")
ax36.imshow(images[5], cmap="gray")
```
